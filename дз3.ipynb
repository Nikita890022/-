```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace (-5,5,25)
y= np.cos(x)+np.random.normal(0,0.1,size=len(x)) #график с шумом
y1=1-x**2/2 + x**4/(4*3*2)- x**6/(6*5*4*3*2)

ax = plt.gca()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

plt.scatter(x,y)
plt.plot(x,y,)
plt.plot(x,y1)
plt.grid(True)
plt.xlabel('Ось х') 
plt.ylabel('Ось y') 
plt.title('График') 
plt.show()
```
