```python
a=int(input())
A=[]

def func(a):
    i=2
    while a > 1:
        while a%i == 0:
            A.append(i)
            a=a//i
        i=i+1
    return A
func(a)
    
```

```python
import numpy as np
x =np.array([0, 1.65, 7.75, 5] , dtype=float)
y =np.array([4.9, 8.301, 20.22, 15.2] ,dtype =float)
sum_x=np.sum(x)
sum_y=np.sum(y)

arr1=x*y
arr2=x*x

sum_x2=np.sum(arr2)
sum_xy=np.sum(arr1)

n=len(x)
def coeff():
    a=(n*sum_xy-sum_x*sum_y)/(n*sum_x2-(sum_x)**2)
    b=((sum_y-a*sum_x))/n
    return (a,b)
a,b=coeff()






import matplotlib.pyplot as plt

def plot_regression(x, y, a, b):
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', label='Данные')
    
    x_line = np.linspace(x.min(), x.max(), 100)
    y_line = a * x_line + b
    plt.plot(x_line, y_line, color='red', label=f'y = {a:.2f}x + {b:.2f}')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Метод наименьших квадратов')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_regression(x, y, a, b)
```
